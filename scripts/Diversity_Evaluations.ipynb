```python
fig, ax1 = plt.subplots(figsize=(12, 8.5))
bottoms = np.zeros(len(df_pivot))

# Stacked Bars
for climate, color in koppen_cls.items():
    if climate in df_pivot.columns:
        ax1.bar(df_pivot.index, df_pivot[climate], bottom=bottoms, color=color)
        bottoms += df_pivot[climate].values

# Overlay Line Plot
ax2 = ax1.twinx()
ax2.plot(df_pivot.index, df_pivot["Total Count"], color='black', marker='o', linestyle='-', linewidth=2,
         markersize=8, label="Total Count")

ax1.set_ylabel("Proportion of Climate Classes Covered", fontsize=14)
ax2.set_ylabel("Unique Locations (millions)", fontsize=14)
ax1.set_xticklabels(df_pivot.index, rotation=40, ha="right", fontsize=13)
ax2.legend(loc="upper left", fontsize=11, bbox_to_anchor=(1.05, -0.03))

# === Legend Axes on the Right ===
legend_ax = ax1.inset_axes([1.06, -0.02, 0.18, 1.06])

legend_ax.axis('off')
legend_ax.set_xlim(0, 3)
legend_ax.set_ylim(-0.5, len(koppen_classes))

# Group category positions
category_bounds = {}
for category, (start_idx, end_idx) in category_labels.items():
    category_bounds[category] = {
        "start": start_idx,
        "end": end_idx,
        "center": (start_idx + end_idx) / 2
    }

# Draw Climate Class Entries
for i, (code, category, color) in enumerate(reversed(koppen_classes)):
    row = len(koppen_classes) - i - 1
    legend_ax.text(0.0, row, code, va='center', ha='left', fontsize=12)#, fontweight='bold')  # class code
    
    # legend_ax.add_patch(plt.Rectangle((0.9, row - 0.35), 0.5, 0.5, color=color, ec='black'))  # color
    legend_ax.add_patch(plt.Rectangle((1.0, row - 0.4), 0.6, 0.6, color=color, ec='black', linewidth=0.5))
    # No need to repeat category here
# legend_ax = ax1.inset_axes([1.07, 0.0, 0.18, 0.99])  # Was 0.15, now 0.18


# Draw Group Dividers + Rotated Labels
for cat, bounds in category_bounds.items():
    # Draw top and bottom lines
    legend_ax.plot([0, 1.6], [bounds["start"] - 0.4, bounds["start"] - 0.4], color="black", linewidth=0.5)
    legend_ax.plot([0, 1.6], [bounds["end"] + 0.6, bounds["end"] + 0.6], color="black", linewidth=0.5)

    # Rotated category label
    legend_ax.text(2.1, bounds["center"], cat, va="center", ha="center",
                   fontsize=14,  rotation=90) #fontweight="bold",


# Adjust layout
fig.subplots_adjust(left=0.07, right=0.9, top=0.95, bottom=0.2)
plt.title("b) Climate Class Proportions Across Datasets with Total Non-Zero Unique Locations", fontsize=16)
# Save
plt.savefig("climate_proportions.pdf", dpi=300, bbox_inches="tight")
# plt.savefig("climate_proportions_rotated_grouped_legend_40.png", dpi=300, bbox_inches="tight")
plt.show()
```
